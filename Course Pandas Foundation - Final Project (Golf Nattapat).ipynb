```python
# TODO 04 - your friend ask for all order data in `California` and `Texas` in 2017 (look at Order Date), send him csv file

# Create year column
df['Order_Year'] = df['Order Date'].dt.year

# filter and export
fil_state_year = df.query("State==['California','Texas'] & Order_Year == 2017" )
fil_state_year.to_csv('fil_state_year.csv')
fil_state_year.head()
```

```python
# TODO 05 - how much total sales, average sales, and standard deviation of sales your company make in 2017
df2017 = df[df['Order_Year'] == 2017]  

df2017['Sales'].agg(['sum','mean','std'])
```

```python
# TODO 06 - which Segment has the highest profit in 2018
df2018 = df[df['Order_Year']==2018]
df2018.groupby('Segment')['Profit'].sum()

# Answer : 'Consumer' segment have highest profit in 2018 (28460.1665 profit)
```

```python
# TODO 07 - which top 5 States have the least total sales between 15 April 2019 - 31 December 2019

# Create month and day column
df['Order_Month'] = df['Order Date'].dt.month
df['Order_Day'] = df['Order Date'].dt.day

#filter
df_fil = df.query("Order_Day >= 15 & Order_Month >= 4 & Order_Year == 2019")

df_fil.groupby(['State'])['Sales'].sum().sort_values().head(5)
```

```python
# TODO 08 - what is the proportion of total sales (%) in West + Central in 2019 e.g. 25% 
sale_tol_2019 = df.query("Order_Year == 2019")['Sales'].sum()
sale_need = df.query("Order_Year == 2019 & Region == ['West','Central']")['Sales'].sum()
x = sale_need / sale_tol_2019

percentage = str(round(x*100)) + '%'
print(percentage)
```

```python
# TODO 09 - find top 10 popular products in terms of number of orders vs. total sales during 2019-2020
# Filter year 2019-2020
fil_year = df.query("Order_Year == [2019,2020]")

# Count number of orders each products
fil_year.groupby(['Product Name'])['Order ID'].count().sort_values(ascending=False).head(10).reset_index()
```

```python
# Sum total sale each products
fil_year.groupby(['Product Name'])['Sales'].sum().sort_values(ascending=False).head(10).reset_index()
```

```python
# TODO 10 - plot at least 2 plots, any plot you think interesting :)

# Top 10 of total profit group by State
df.groupby(['State'])['Profit'].sum().sort_values(ascending=False).head(10).plot(kind='bar');
```

```python
# Which month most sale
df.groupby(['Order_Month'])['Sales'].sum().plot(kind='line');
```
